```python
import pandas as pd
from dateutil.relativedelta import relativedelta
from datetime import datetime, date
```

* Função para calcular a idade

```python
# This function converts given date to age
def age(born,year):
    born = datetime.strptime(born, "%d-%m-%Y").date()
    today = datetime.strptime(f"01-01-{year}", "%d-%m-%Y").date()
    return today.year - born.year - ((today.month, 
                                      today.day) < (born.month, 
                                                    born.day))
```

# Tratamento dos dados das planilhas de 2015 à 2017
## Cálculo da Idade

```python
df_2015 = pd.read_csv("PDA_PROUNI_2015_CSV.csv", delimiter=";", encoding='windows-1252')
```

```python
# Calcula Idade
df_2015['IDADE'] = df_2015['DT_NASCIMENTO_BENEFICIARIO'].apply(lambda x: age(x,2015))
```

```python
df_2015.head()
```

```python
df_2016 = pd.read_csv("PDA_PROUNI_2016_CSV.csv", delimiter=";", encoding='windows-1252')
```

```python
df_2016['IDADE'] = df_2016['DT_NASCIMENTO_BENEFICIARIO'].apply(lambda x: age(x,2016))
```

```python
df_2016.head()
```

```python
df_2017 = pd.read_csv("pda-bolsistas-prouni-2017.CSV", delimiter=";", encoding='windows-1252')
```

```python
df_2017.head()
```

### Foi necessário converter o formato da data na planilha de 2017

```python
# Converte formato da data
df_2017["DT_NASCIMENTO_BENEFICIARIO"] = pd.to_datetime(df_2017["DT_NASCIMENTO_BENEFICIARIO"])
df_2017["DT_NASCIMENTO_BENEFICIARIO"] = df_2017["DT_NASCIMENTO_BENEFICIARIO"].dt.strftime("%d-%m-%Y")
df_2017.head()
```

```python
df_2017['IDADE'] = df_2017['DT_NASCIMENTO_BENEFICIARIO'].apply(lambda x: age(x,2017))
```

```python
df_2017.head()
```

## Junta os dataframes

```python
# Junta DFs
df = pd.concat([df_2015,df_2016,df_2017])
```

```python
df.head()
```

```python
len(df)
```

## Filtra pelo ano entre 2015 e 2017

```python
# Filtra apenas pelos anos corretos
df = df[(df['ANO_CONCESSAO_BOLSA']>=2015) & (df['ANO_CONCESSAO_BOLSA']<=2017)]
```

```python
len(df)
```

```python
df['ANO_CONCESSAO_BOLSA'].unique()
```

## Filtra pela idade

```python
df['IDADE'].unique()
```

```python
# Filtra pela idade
df = df[df['IDADE']>15]
```

```python
len(df)
```

## Cria uma coluna com a faixa de idade (< 25, >=25 e <=40, > 40)

```python
def calcula_faixa_idade(age):
    if age < 25:
        return '< 25'
    if age >= 25 and age <= 40:
        return '>= 25 e <= 40'
    if age > 40:
        return '> 40'
```

```python
# Cria faixa de idade
df['FAIXA_IDADE'] = df['IDADE'].apply(calcula_faixa_idade)
```

```python
df['FAIXA_IDADE'].unique()
```

```python
df['FAIXA_IDADE'].nunique()
```

```python
df.head()
```

```python
df.dtypes
```

## Padroniza colunas do tipo texto, colocando a primeira letra de cada palavra em maísculo

```python
names_columns = ['NOME_IES_BOLSA','TIPO_BOLSA','MODALIDADE_ENSINO_BOLSA','NOME_CURSO_BOLSA','NOME_TURNO_CURSO_BOLSA','SEXO_BENEFICIARIO_BOLSA','RACA_BENEFICIARIO_BOLSA','BENEFICIARIO_DEFICIENTE_FISICO','MUNICIPIO_BENEFICIARIO_BOLSA']
```

```python
# Padroniza colunas strings
for column in names_columns:
    df[column] = df[column].apply(lambda x: str(x).title())
```

```python
df.head()
```

## Padroniza colunas região e UF, colocando tudo em maísculo

```python
# Padroniza colunas string
names_columns = ['REGIAO_BENEFICIARIO_BOLSA', 'SIGLA_UF_BENEFICIARIO_BOLSA']
for column in names_columns:
    df[column] = df[column].apply(lambda x: str(x).upper())
```

```python
df.head()
```

## Cria um índice único para cada linha

```python
# Reseta index
df = df.reset_index()
```

```python
df = df.drop('index',axis=1)
```

## Salva em CSV

```python
df.to_csv('prouni.csv')
```

# Calculo de bolsas per capita no ano de 2015
## Carrega planilha com dados da população dos estados em 2015 e realiza alguns tratamentos

```python
df_populacao_estados = pd.read_csv('populacao_estados_2015.csv', delimiter=";")
```

```python
df_populacao_estados.head()
```

```python
# Remove linha do Brasil
df_populacao_estados = df_populacao_estados.iloc[1:,:]
```

```python
df_populacao_estados.head()
```

```python
# Renomeia colunas
df_populacao_estados = df_populacao_estados.rename(columns={'Cód.':'Cod','Brasil e Unidade da Federação':'Total'})
```

```python
df_populacao_estados['Cod'].unique()
```

```python
states = {
    'Acre': 'AC',
    'Alagoas': 'AL',
    'Amapá': 'AP',
    'Amazonas': 'AM',
    'Bahia': 'BA',
    'Ceará': 'CE',
    'Distrito Federal': 'DF',
    'Espírito Santo': 'ES',
    'Goiás': 'GO',
    'Maranhão': 'MA',
    'Mato Grosso': 'MT',
    'Mato Grosso do Sul': 'MS',
    'Minas Gerais': 'MG',
    'Pará': 'PA',
    'Paraíba': 'PB',
    'Paraná': 'PR',
    'Pernambuco': 'PE',
    'Piauí': 'PI',
    'Rio de Janeiro': 'RJ',
    'Rio Grande do Norte': 'RN',
    'Rio Grande do Sul': 'RS',
    'Rondônia': 'RO',
    'Roraima': 'RR',
    'Santa Catarina': 'SC' ,
    'São Paulo': 'SP',
    'Sergipe': 'SE',
    'Tocantins': 'TO' 
}
```

```python
# Converte estados para siglas
df_populacao_estados['UF'] = df_populacao_estados['Cod'].apply(lambda x: states[x])
```

```python
df_populacao_estados.head()
```

```python
# Filtra colunas
df_populacao_estados = df_populacao_estados[['UF','Total']]
```

```python
df_populacao_estados.head()
```

```python
# Sigla do estado como índice
df_populacao_estados = df_populacao_estados.set_index('UF')
```

```python
df_populacao_estados.head()
```

## Realiza o cálculo de bolsas per capita

```python
# DF para calcular bolsas per capita por estado em 2015
df_per_capita = df[df['ANO_CONCESSAO_BOLSA']==2015]
```

```python
grouped = df_per_capita.groupby(['SIGLA_UF_BENEFICIARIO_BOLSA']).size()
```

```python
grouped_df = pd.DataFrame(grouped,columns=['TotalBolsas'])
```

```python
grouped_df.head()
```

```python
# Join com tabela com total da população do estado
joined = grouped_df.join(df_populacao_estados,how="inner")
```

```python
joined.head()
```

```python
joined['PER_CAPITA'] = joined['TotalBolsas']/joined['Total']
```

```python
# Calcula per capita
joined.head()
```

```python
joined.sort_values('PER_CAPITA',ascending=False)
```

```python
joined = joined.drop(['TotalBolsas','Total'],axis=1)
joined.head()
```

## Salva em CSV

```python
joined.to_csv('bolsas_per_capita_2015.csv')
```
